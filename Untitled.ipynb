```python
import seppopy.tools.db as db
```

```python
conn,cur = db.connect(dbname='ebdqueues',user='ebdqueueuser')
```

```python
sql="select column_name from information_schema.columns where table_name='queues'"
```

```python
cur.execute(sql)
res=cur.fetchall()
conn.commit()
```

```python
column_names = [x[0] for x in res]
column_names
```

```python
sql='select * from queues'
cur.execute(sql)
res=cur.fetchall()
conn.commit()
print(len(res))
```

```python
# Build a list of all queues as dictlist :
ebdqueues = []
for i in res:
    ebdqueues.append(dict(zip(column_names,i)))
```
